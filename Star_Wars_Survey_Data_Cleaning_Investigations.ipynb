# In this project we explore data from the Star Wars Survey from FiveThirtyEight

```python
# read in the data
import pandas as pd
star_wars = pd.read_csv("star_wars.csv",encoding="ISO-8859-1")
```

```python
# take a brief look at the data (first ten rows)
star_wars.head(10)
```

### It looks like there are some strange values when the respondent's ID is missing or when the survey respondent did not see any of the Star Wars movies.

```python
# remove rows where RespondentID is NaN
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head()
```

```python
# convert the "Have you seen any of the 6 films in the Star Wars 
# franchise?" and "Do you consider yourself to be a fan of the Star
# Wars film franchise?" columns to Boolean types

# get unique values
print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts())
print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts())

# convert using a dictionary:
yes_no = {"Yes":True,
          "No":False}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

# check
print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts())
print(star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts())
```

### Clean up some of the survey question data

```python
# convert the "Which of the following Star Wars films have you seen? 
# Please select all that apply." and "Unnamed: 4" - "Unnamed: 8"
# columns so that they contain Boolean values
import numpy as np
for i in range(3,9):
    print(star_wars.iloc[:,i].value_counts())
dict = {"Star Wars: Episode I  The Phantom Menace":True,
        "Star Wars: Episode II  Attack of the Clones":True,
        "Star Wars: Episode III  Revenge of the Sith":True,
        "Star Wars: Episode IV  A New Hope":True,
        "Star Wars: Episode V The Empire Strikes Back":True,
        "Star Wars: Episode VI Return of the Jedi":True,
        np.nan:False}
for i in range(3,9):
    star_wars.iloc[:,i] = star_wars.iloc[:,i].map(dict)
for i in range(3,9):
    print(star_wars.iloc[:,i].value_counts())
star_wars.head()
```

```python
# rename these columns to have more intuitive names
col_dict = {}
num = 0
for i in range(3,9):
    num += 1
    col_dict[star_wars.columns[i]]="seen_"+str(num)
star_wars = star_wars.rename(columns=col_dict)

# check
star_wars.columns
```

### Clean up the next set of columns which represent rankings of the 6 Star Wars movies

```python
# convert columns 9 - 14 to float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# rename columns
col_dict = {}
num = 0
for i in range(9,15):
    num += 1
    col_dict[star_wars.columns[i]]="ranking_"+str(num)
star_wars = star_wars.rename(columns=col_dict)

# check
star_wars.iloc[:,9:15].head()
```

### Find the highest-ranked Star Wars movie

```python
# find the mean value of the ranking columns
means = star_wars.iloc[:,9:15].mean()
```

```python
# make a bar chart
%matplotlib inline
means.plot.bar()
```

### Looks like the Star Wars movie with the highest ranking is Star Wars: Episode V The Empire Strikes Back (the one with the iconic light saber battle between Luke and Darth Vader).

### Also get the number of respondents who saw each movie:

```python
num_saw_movie = star_wars.iloc[:,3:9].sum()
num_saw_movie.plot.bar()
print(star_wars.shape)
```

### I looks like the favorite movie (Episode 5) had the most views (note the total number of possible views in this dataset is 1186).  Also the original trilogy had more views, with the views of each subsequent prequel falling off (probably because of Jar Jar Binks--ugh).

### Re-examine the movie rankings and views after splitting on gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
df = pd.DataFrame()
df["males"] = males.iloc[:,9:15].mean()
df["females"] = females.iloc[:,9:15].mean()
df.plot.bar()
print(df)
```

### Looks like the general ranking trend is pretty similar and Episode V is still the favorite, but the overall rankings are slightly different.  From lowest to highest for males the ranking is 5, 4, 6, 1, 2, 3; while for females it is 5, 6, 1, 4, 2, 3.

```python
num_saw_movie = pd.DataFrame()
num_saw_movie["males"] = males.iloc[:,3:9].sum()
num_saw_movie["females"] = females.iloc[:,3:9].sum()
num_saw_movie.plot.bar()
```

### Again the overall pattern in the number of respondents who viewed the different Star Wars movies is very similar.  However, there are less views from female respondents.

```python
# check overall numbers of male and female respondents
print(males.shape[0])
print(females.shape[0])
```
