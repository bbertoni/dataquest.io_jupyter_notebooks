# Visualizing the Gender Gap in College Degrees

### In this notebook, we use gender data from college graduates (from a dataset compiled by Randal Olson) to visualize the gender cap in different college degrees.

We plot the data, group it by degree category, hide the x-axis labels, set conservative y-axis labels, and add a horizontal line to ease a high-level visualization of the gender gap in various college degrees

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig = plt.figure(figsize=(18, 21))

def make_plot(sp,index_val):
    ax = fig.add_subplot(6,3,index_val)
    ax.plot(women_degrees['Year'], x, c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-x, c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(label)
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255),alpha=0.3)
    if sp == 0:
        ax.text(2004, 83, 'Women')
        ax.text(2005, 12, 'Men')
    elif sp == 15:
        ax.text(2005, 87, 'Men')
        ax.text(2004, 8, 'Women')
        ax.tick_params(labelbottom="on")
    elif sp == 1:
        ax.text(2004, 76, 'Women')
        ax.text(2005, 20, 'Men')
    elif sp == 2:
        ax.text(2004, 91, 'Women')
        ax.text(2005, 6, 'Men')
    elif sp == 17:
        ax.text(2005, 63, 'Men')
        ax.text(2004, 32, 'Women')
        ax.tick_params(labelbottom="on")
    elif sp == 13:
        ax.tick_params(labelbottom="on")
        
i = 0
j = 0
k = 0

for sp in range(0,18):
    if sp % 3 == 0:
        x = women_degrees[stem_cats[i]]
        label = stem_cats[i]
        make_plot(sp=sp,index_val=sp+1)
        i += 1
    elif sp % 3 == 1 and sp != 16:
        x = women_degrees[lib_arts_cats[j]]
        label = lib_arts_cats[j]
        make_plot(sp=sp,index_val=sp+1)
        j += 1
    elif sp % 3 == 2:
        x = women_degrees[other_cats[k]]
        label = other_cats[k]
        make_plot(sp=sp,index_val=sp+1)
        k += 1
    else:
        pass

plt.savefig('gender_gap_college_degrees.png')    
plt.show()
```
