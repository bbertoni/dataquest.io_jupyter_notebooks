# Exploring Data Visualizations in Pandas

```python
# import pandas and matplotlib
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# read dataset into a dataframe and take a look
# at the data
recent_grads = pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

```python
# top five majors by median earnings
# guess I should've majored in engineering...
recent_grads.head()
```

```python
# bottom five majors by median earnings
recent_grads.tail()
```

```python
# summary statistics
recent_grads.describe()
```

```python
# drop rows with missing values
raw_data_count = recent_grads.shape[0]
raw_data_count
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
cleaned_data_count
```

```python
# explore the data using some scatter plots
# this plot shows that the more populat majors do
# not appear to make more money, however these
# conclusions could be limited due to the small
# sample size
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(5,5))
```

```python
# this plot shows that (outside of the scatter
# for the majors with low statistics), the 
# unemployment rate is roughly constant
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment rate vs. Sample_size', figsize=(5,5))
```

```python
# this shows that the median salary seems to
# decrease with the number of full time
# employees
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full Time', figsize=(5,5))
```

```python
# this plot shows that there is not really any
# correlation between the unemployment rate for
# a major and the share of women in that major
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs. Share of Women', figsize=(5,5))
```

```python
# this plot shows that the median salary is
# fairly constant/slightly decreases with the 
# number of male graduates
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men', figsize=(5,5))
```

```python
# the plot is similar for female graduates
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Women', figsize=(5,5))
```

```python
# explore the data using histograms
# from this plot we can see the small sample size 
# problem
recent_grads["Sample_size"].hist(bins=30,range=(0,3000))
plt.title("Sample Size")
```

```python
# from this plot we can see that typical salaries
# are around $30K to $40K for college graduates
recent_grads['Median'].hist()
plt.title('Median Salary')
```

```python
# employment numbers look small, but not when
# compared to the sample size
recent_grads['Employed'].hist(bins=20,range=(0,200000))
plt.title('Number employed')
```

```python
# same goes for this full time employment 
# histogram
recent_grads['Full_time'].hist(bins=20,range=(0,200000))
plt.title('Number employed full time')
```

```python
# this histogram shows that most majors have more
# females compared to males
recent_grads['ShareWomen'].hist()
plt.title('Women as share of total')
```

```python
# the unemployment rate looks reasonably
# normally distributed, with a mean around 0.7
recent_grads['Unemployment_rate'].hist()
plt.title('Unemployment rate')
```

```python
# most majors contain only less than 10K male
# graduates
recent_grads['Men'].hist(bins=20,range=(0,120000))
plt.title('Number of male graduates')
```

```python
# similarly most majors have less than 20K 
# female graduates
recent_grads['Women'].hist(bins=30,range=(0,200000))
plt.title('Number of women graduates')
```

```python
# try making scatterplot matrices
from pandas.plotting import scatter_matrix
# this plot shows that there is a lot of scatter
# in the median salary for small sample sizes
# and the median salary is somewhat constant
# around about $40K for larger sample sizes
scatter_matrix(recent_grads[["Sample_size","Median"]])
```

```python
# this plot shows that there is not much
# correlation between the median salary for a
# major and the unemployment rate.  We also see
# a lot of scatter in the unemployment rate for
# the majors with a small sample size, but for
# larger sample sizes the unemployment rate
# flattens around 0.7
scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']])
```

```python
# explore the data using bar plots
# the top ten majors by median salary are male-
# dominated
# hurray for women in astrophysics though!
recent_grads[:10].plot.bar(x='Major',y='ShareWomen')
plt.title('Share of women in the top ten majors by median salary')
```

```python
# sigh...these majors are mostly women
recent_grads[-10:].plot.bar(x='Major',y='ShareWomen')
plt.title('Share of women in the bottom ten majors by median salary')
```

```python
# surprisingly the unemployment rates are 
# sometimes quite high for the top ten majors
# with highest median salary
# The unemployment rate for petroleum engineering
# and astrophysics/astronomy is quite low though!
recent_grads[:10].plot.bar(x='Major',y='Unemployment_rate')
plt.title('Unemployment rate in the top ten majors by median salary')
```

```python
# unemployment rates for the 10 majors with the
# lowest median salaries look a little higher
# than the average 0.7
recent_grads[-10:].plot.bar(x='Major',y='Unemployment_rate')
plt.title('Unemployment rate in the bottom ten majors by median salary')
```
